# Datos adicionales a ser agregados en DB


Se hará un `merge` del dataset maestro `20210823_Región_de_Aysén_del_General_Carlos_Ibáñez_del_Campo_notificaciones_0600.csv` y el diccionario de provincias y comunas`establecimientos_salud_dict.xls` creado por Miguel Oyarzo

```python
import pandas as pd
```

```python
df = pd.read_csv('data/20210826_Región de Aysén del General Carlos Ibáñez del Campo_notificaciones_0600.csv', 
                   sep=";", # CSV con separador de semicolon
                   low_memory=False
                  )
```

```python
df.sample(5)
```

```python
df2 = pd.read_excel('data/establecimientos_salud_dict.xlsx')
```

```python
df2.sample(5)
```

```python
data_con_establecimiento_cp = df.merge(df2)
```

## Integra datos poblacionales

Se realizará merge del dataset recientemente generado con el diccionario de estimación poblacional


```python
# Importa datos poblacionales

df_pop = pd.read_csv('data/aysen_population.csv')
```

```python
df_pop
```

```python
df = data_con_establecimiento_cp.merge(df_pop.drop(columns= ['Unnamed: 0','orden','nombre_region', 'codigo_region', 'nombre_provincia','codigo_provincia']), left_on='comuna_seremi_xi', right_on='nombre_comuna' )
```

```python
df.sample(5)
```

```python
# Exporta DB editada con datos de población

df.to_csv('data/df_clean.csv')
```
