```python
df_pop = df_pop.clean_names().rename(columns={'hombres':'edad_hombres','mujeres':'edad_mujeres'}).drop(columns=['edad'])

df_pop
```

```python
# Valores deben ser transformados de formato upper a title 

df_pop['nombre_region'] = df_pop['nombre_region'].str.title()

df_pop['nombre_region'] = 'Región de Aysén del General Carlos Ibáñez del Campo'

df_pop['nombre_provincia'] = df_pop['nombre_provincia'].str.title()

df_pop['nombre_comuna'] = df_pop['nombre_comuna'].str.title()

df_pop
```

```python
# Debemos eliminar tildes de los valores evitando la remoción de la virgulilla de la ñ

import re
from unicodedata import normalize

def tilde_remover(x):
    return normalize( 'NFC', re.sub(
                r"([^n\u0300-\u036f]|n(?!\u0303(?![\u0300-\u036f])))[\u0300-\u036f]+", r"\1", 
                normalize( "NFD", x), 0, re.I
            ))
```

```python
df_pop['nombre_provincia'] = df_pop['nombre_provincia'].apply(lambda x: tilde_remover(x))

df_pop['nombre_comuna'] = df_pop['nombre_comuna'].apply(lambda x: tilde_remover(x))
```
